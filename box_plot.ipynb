```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Données simulées (moyenne et écart-type globales par année)
global_stats = {
    2015: {'mean': 2, 'stddev': 2.828427125},
    2018: {'mean': 2, 'stddev': 3},
    2019: {'mean': 2, 'stddev': 3},
    2020: {'mean': 2, 'stddev': 3.16227766}
}

# Données de vos brevets
data = {
    'appln_filing_year': [2015, 2018, 2018, 2019, 2019, 2019, 2019, 2019, 2019, 2020],
    'normalise': [0.707106781, 0.333333333, 0, -0.333333333, 0.666666667, 0, -0.333333333, 0, 0.666666667, -0.316227766]
}
df = pd.DataFrame(data)

# Création du graphique
plt.figure(figsize=(10, 6))

# Tracer les boxplots pour chaque année
for year in global_stats:
    # Simuler une distribution normale pour chaque année
    mean = global_stats[year]['mean']
    stddev = global_stats[year]['stddev']
    simulated_data = np.random.normal(mean, stddev, 1000)

    # Créer un boxplot pour la distribution simulée
    plt.boxplot(
        simulated_data,
        positions=[year],
        widths=0.6,
        patch_artist=True,
        showfliers=False,
        boxprops=dict(facecolor="skyblue", alpha=0.6)  # Ajuster l'opacité pour laisser apparaître les points
    )

# Ajouter les scores normalisés de vos brevets comme des points rouges
for i, row in df.iterrows():
    plt.scatter(
        row['appln_filing_year'], row['normalise'],
        color='red',
        zorder=5,  # Met les points au premier plan
        label='Vos brevets' if i == 0 else ""
    )

# Ajout des labels et titre
plt.title("Comparaison des scores normalisés des brevets avec la distribution globale")
plt.xlabel("Année de dépôt")
plt.ylabel("Score normalisé")
plt.xticks(list(global_stats.keys()))
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Calculer la moyenne des scores normalisés par année
average_normalized = df.groupby('appln_filing_year')['normalise'].mean()

# Afficher les résultats
print(average_normalized)

# Visualisation
average_normalized.plot(kind='bar', figsize=(8, 6), color='skyblue', edgecolor='black')
plt.title('Moyenne des scores normalisés par année')
plt.xlabel('Année de dépôt')
plt.ylabel('Moyenne des scores normalisés')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['appln_filing_year'], df['normalise'], marker='o', linestyle='-', color='blue')
plt.title('Évolution des scores normalisés des brevets')
plt.xlabel('Année de dépôt')
plt.ylabel('Score normalisé')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Données des demandes
demandes = {'EP': 4, 'FR': 5, 'KR': 1, 'WO': 2}  # Total des secteurs demandés
accordes = {'EP': 2, 'FR': 4, 'KR': 0, 'WO': 0}  # Total des secteurs accordés

# Conversion en DataFrame pour faciliter le tracé
df = pd.DataFrame([demandes, accordes], index=['Demandes', 'Accordés']).T

# Tracé de l'histogramme
df.plot(kind='bar', figsize=(8, 6), color=['skyblue', 'lightgreen'], edgecolor='black')

# Ajouter des labels et un titre
plt.title('Comparaison entre brevets demandés et accordés par office')
plt.xlabel('Office de dépôt')
plt.ylabel('Nombre de secteurs technologiques')
plt.xticks(rotation=0)
plt.legend(title='Statut')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Affichage du graphique
plt.show()
```
